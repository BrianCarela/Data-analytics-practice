## 2. The Four Libraries — Quick Reference

| Library | Nickname | What it does | When to use it |
|---|---|---|---|
| `numpy` | `np` | Fast math on arrays | Stats, array math, working under the hood |


## 3. Python Syntax Cheat Sheet

### Variables & Types
```python
name = "Alice"          # str
age = 34                # int
bmi = 27.4              # float
is_smoker = False       # bool (capital T/F)
missing = None          # NoneType (like null in JS)

# Type conversion
int("42")               # str → int
float("3.14")           # str → float
str(42)                 # int → str
"yes" == "yes"          # → True (boolean from comparison)

# f-strings
f"{name}, age {age}"                    # basic
f"${amount:,.2f}"                       # comma + 2 decimals
f"{value:.1f}"                          # 1 decimal place
f"{'Yes' if is_smoker else 'No'}"       # inline if/else
```

### Lists
```python
nums = [1, 2, 3, 4, 5]

nums[0]          # first item → 1
nums[-1]         # last item  → 5
nums[1:3]        # slice      → [2, 3]  (start inclusive, stop exclusive)
nums[:3]         # first 3   → [1, 2, 3]
nums[-2:]        # last 2    → [4, 5]
nums[::-1]       # reversed  → [5, 4, 3, 2, 1]

nums.append(6)   # add to end
nums.insert(0, 0) # insert at index
nums.remove(3)   # remove first occurrence of value
nums.pop()       # remove & return last item
nums.pop(0)      # remove & return item at index 0

len(nums)        # count
sum(nums)        # total
min(nums)        # smallest
max(nums)        # largest
sorted(nums)     # new sorted list (original unchanged)
sorted(nums, reverse=True)  # descending
5 in nums        # True/False membership check
```

### Loops
```python
# Basic for loop
for item in collection:
    print(item)

# With index
for i, item in enumerate(collection):
    print(i, item)

# Two lists at once
for a, b in zip(list1, list2):
    print(a, b)

# Range
for i in range(5):          # 0 1 2 3 4
for i in range(1, 6):       # 1 2 3 4 5
for i in range(0, 10, 2):   # 0 2 4 6 8

# Accumulator pattern
results = []
for item in collection:
    if condition:
        results.append(item)
```

### List Comprehensions
```python
# Transform every item
[x * 2 for x in nums]

# Filter
[x for x in nums if x > 0]

# Transform + filter
[x * 2 for x in nums if x > 0]

# Inline if/else (no filter — assigns one of two values)
['High' if x > 100 else 'Low' for x in nums]

# With zip
[f"{n}: {c}" for n, c in zip(names, charges)]

# Apply a function
[classify_bmi(b) for b in bmis]
```

### Functions
```python
def function_name(param1, param2, optional=default_value):
    """Docstring describing what this does."""
    result = param1 + param2
    return result

# Call by position
function_name(10, 20)

# Call by keyword (order doesn't matter)
function_name(param2=20, param1=10)

# Return multiple values (returns a tuple)
def stats(nums):
    return min(nums), max(nums), sum(nums)/len(nums)

low, high, avg = stats(nums)  # unpack
```

## 4. Pandas Quick Reference

### Loading Data
```python
df = pd.read_csv('file.csv')                  # from local file
df = pd.read_csv('https://...')               # from URL
df = pd.DataFrame(data_dict)                  # from Python dict
df = sns.load_dataset('tips')                 # seaborn built-in datasets
```

### First Look (always run these first)
```python
df.shape          # (rows, columns)
df.head()         # first 5 rows
df.head(10)       # first 10 rows
df.tail()         # last 5 rows
df.info()         # column names, types, non-null counts
df.describe()     # summary stats for numeric columns
df.isnull().sum() # count of missing values per column
df.dtypes         # data type of each column
df.columns        # list of column names
```

### Selecting Data
```python
df['col']                    # single column → Series
df[['col1', 'col2']]         # multiple columns → DataFrame
df.iloc[0]                   # row by integer position
df.loc[0]                    # row by label/index
df.iloc[0:5]                 # first 5 rows
```

### Filtering Rows
```python
df[df['col'] == 'value']              # equal to
df[df['col'] > 100]                   # greater than
df[df['col'].isin(['a', 'b'])]        # in a list
df[df['col'].notna()]                 # not missing

# Multiple conditions — use & (AND) | (OR), wrap each in ()
df[(df['age'] > 30) & (df['smoker'] == 'yes')]
df[(df['region'] == 'northeast') | (df['region'] == 'northwest')]
```

### Series Stats
```python
df['col'].mean()      # average
df['col'].median()    # median
df['col'].std()       # standard deviation
df['col'].min()       # minimum
df['col'].max()       # maximum
df['col'].sum()       # total
df['col'].count()     # non-null count
df['col'].value_counts()  # frequency of each unique value
df['col'].nunique()   # number of unique values
df['col'].unique()    # array of unique values
```

### GroupBy
```python
# Single group, single stat
df.groupby('region')['charges'].mean()

# Single group, multiple stats
df.groupby('smoker')['charges'].agg(['mean', 'median', 'count'])

# Two groups (cross-tab)
df.groupby(['region', 'smoker'])['charges'].mean().unstack()

# Sort result
df.groupby('region')['charges'].mean().sort_values(ascending=False)
```

### Adding & Modifying Columns
```python
df['new_col'] = df['col'] * 1.1              # math on existing column
df['category'] = df['col'].apply(my_func)    # apply a function row by row
df['flag'] = df['col'] > 100                 # boolean column
df['col'] = df['col'].astype(float)          # change data type
df['col'] = df['col'].str.lower()            # string operations
df['col'] = df['col'].fillna(0)              # fill missing values
df = df.drop(columns=['col'])                # remove a column
df = df.rename(columns={'old': 'new'})       # rename a column
```

### Sorting
```python
df.sort_values('charges')                          # ascending
df.sort_values('charges', ascending=False)         # descending
df.sort_values(['region', 'charges'])              # multi-column sort
```

## 5. NumPy Quick Reference

```python
import numpy as np

# Create arrays
arr = np.array([1, 2, 3, 4, 5])
arr = df['col'].values          # pandas Series → numpy array

# Math (applies to whole array at once)
arr * 2
arr + 100
arr ** 2                        # squared
arr > 50                        # boolean array
arr[arr > 50]                   # filter using boolean array

# Stats
np.mean(arr)                    # average
np.median(arr)                  # median
np.std(arr)                     # standard deviation
np.min(arr)                     # minimum
np.max(arr)                     # maximum
np.sum(arr)                     # total
np.percentile(arr, 25)          # 25th percentile
np.percentile(arr, 75)          # 75th percentile

# IQR (spread of the middle 50%)
iqr = np.percentile(arr, 75) - np.percentile(arr, 25)
```

### Mean vs Median — when to use which
| Situation | Use |
|---|---|
| Symmetric distribution (no outliers) | Mean |
| Skewed distribution or outliers present | Median |
| Salary, charges, prices, housing | Usually Median |
| Test scores, temperatures | Usually Mean |

**Rule of thumb:** if mean > median, data is right-skewed (long tail on the right). If mean < median, left-skewed.

## 6. Common Errors & Fixes

### `TypeError: Could not convert string to numeric`
**Cause:** You're calling `.mean()` or similar on a column that pandas thinks is text.
**Fix:** Check `df.dtypes`. If the column shows `object` instead of `int64` or `float64`, convert it:
```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```
Also check: did the CSV header row get loaded as a data row? Shape will be off by 1.

---

### `KeyError: 'column_name'`
**Cause:** The column name doesn't exist in the DataFrame.
**Fix:** Print `df.columns` to see exact names. Watch for extra spaces, wrong case.
```python
print(df.columns.tolist())
```

---

### `NameError: name 'df' is not defined`
**Cause:** The cell that creates `df` hasn't been run yet.
**Fix:** Run cells top to bottom. Use Run All if you've reopened the notebook.

---

### `ValueError: operands could not be broadcast together`
**Cause:** Two arrays/columns have different lengths.
**Fix:** Check `len()` of both. Usually happens when a filter creates a shorter Series.

---

### Chart shows but looks wrong / empty
**Cause:** Filtering reduced data to 0 rows, or column has wrong dtype.
**Fix:** Print the filtered DataFrame before charting to verify it has data.

---

### `IndentationError`
**Cause:** Mixed spaces and tabs, or wrong indent level.
**Fix:** Use 4 spaces consistently. Most editors do this automatically.

---

### Using `and`/`or` instead of `&`/`|` in pandas filters
**Cause:** `and`/`or` don't work on Series — they're for single booleans.
**Fix:** Always use `&` and `|` in DataFrame filters, and wrap each condition in `()`:
```python
# Wrong
df[df['age'] > 30 and df['smoker'] == 'yes']

# Right
df[(df['age'] > 30) & (df['smoker'] == 'yes')]
```
